## Customizing Plots created with Pandas

```python
import pandas as pd
```

```python
df2.plot.line()
```

```python
df2['c'].plot.line()
```

```python
df2['c'].plot.line(figsize=(12,8), ls=':')
```

```python
df2['c'].plot.line(figsize=(12,8), ls=':',c='red',lw=5)
```

```python
title = "MY PLOT TITLE"
xlabel = "My X Data"
ylabel = "My Y Data"
```

```python
ax = df2['c'].plot.line(figsize=(12,8), ls=':',c='red',lw=5)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
ax = df2['c'].plot.line(figsize=(12,8), ls=':',c='red',lw=5, title=title)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
df2.plot()
```

```python
ax = df2.plot()
ax.legend(loc=0)
```

```python
ax = df2.plot()
ax.legend(loc=0,bbox_to_anchor=(2.0,1.0))
```

```python
ax = df2.plot()
ax.legend(loc=0,bbox_to_anchor=(1.05,0.3))
```
